```python
import pandas as pd
import os
import datetime as dt
import numpy as np
```

```python
def feasible_patients(df, cat='Vitalwert', days_interval=5):
    patient_list = sorted(df.FallNr.unique())
    ident_list = sorted(df[df.Kategorie==cat].Wertbezeichner.unique())
    test = []
    all_dates = patient_list[:]
    for pat in patient_list:
        df_pat = df[(df.FallNr==pat)&(df.Kategorie==cat)]
        df_pat_temp = df_pat.groupby([df_pat['Zeitstempel'].dt.date, df_pat['Wertbezeichner']]).mean()
        first_date = list(df_pat['Zeitstempel'].dt.date)[0] 
        last_date = first_date + dt.timedelta(days=days_interval-1)
        date_list = [first_date + dt.timedelta(days=x) for x in range(days_interval)]
        index_list = []
        for i in range(len(df_pat_temp[first_date:last_date].index)):
            index_list.append(df_pat_temp.index[i][0])
        diff = list(set(sorted(list(set(index_list)))).symmetric_difference(set(sorted(date_list))))
        if len(diff)>=1:
            all_dates.remove(pat)
    return all_dates
```
